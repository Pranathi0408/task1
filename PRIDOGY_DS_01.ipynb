```python
#plotting bar chart for distribution of region
Region_counts = data['Region'].value_counts() #calculate the all the indivials values of data set
print(f'\ncounts of countries by Region  :\n\n{Region_counts}\n')
x=range(len(Region_counts.index))

#coustimizing the plot
plt.figure(figsize=(10, 6))
plt.bar(Region_counts.index,Region_counts.values,color='#CD1FD2')
plt.xlabel('Region groups', fontsize=15)
plt.ylabel('Region Counts', fontsize=15)
plt.title('Categorical Distribution of Region',fontsize=18)

plt.xticks(x,Region_counts.index,rotation=50,color='#1FBCD2', fontsize=16)
plt.legend(title='Region',fontsize=15)
plt.show()
```

```python
#plotting bar chart for distribution of income
IncomeGroup_counts = data['IncomeGroup'].value_counts() #calculate the all the indivials values of data set
print(f'\ncounts of countries by IncomeGroup  :\n\n{IncomeGroup_counts}\n')
x=range(len(Region_counts.index))

#coustimizing the plot
plt.figure(figsize=(10, 6))
plt.bar(IncomeGroup_counts.index,IncomeGroup_counts.values,color='#D21F1F')
plt.xlabel('Income Groups', fontsize=15)
plt.ylabel('Counts', fontsize=15)
plt.title('Categorical Distribution of IncomeGroups',fontsize=18)
plt.legend(title='IncomeGroups',fontsize=15)
plt.show()#to show the graph
```

```python
#Creating a 2D catogorical distribution with Region and IncomeGroup
categorical_distribution = pd.crosstab(data['Region'],data['IncomeGroup'])
#plotting a bar plot using pandas plot() funtion
categorical_distribution.plot(kind='bar',stacked = True,figsize=(10, 8), fontsize=12)
#coustimizing the plot (optional)
plt.xlabel('Region', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('2-D Categorical Distribution: Region vs IncomeGroups',fontsize=15)
plt.legend(title='IncomeGroups',fontsize=15)
plt.tight_layout
plt.show()#to show the graph
```
